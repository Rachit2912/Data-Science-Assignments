```python
### Q1. Explain GET and POST methods.
'''
- **GET**: Used to request data from the server. It sends parameters through the URL and is typically used for retrieving data.
- **POST**: Used to submit data to the server. Data is sent in the request body, making it suitable for sending sensitive or large data like forms.

---

### Q2. Why is `request` used in Flask?

The `request` object in Flask is used to access incoming request data, such as form data, query parameters, JSON data, or headers. It helps in handling user input and processing HTTP requests.

---

### Q3. Why is `redirect()` used in Flask?

The `redirect()` function in Flask is used to redirect the user to a different URL or route. It's commonly used after form submission to redirect users to a success page or another route.

---

### Q4. What are templates in Flask? Why is the `render_template()` function used?

**Templates** in Flask are HTML files used to render dynamic content in web pages. The **`render_template()`** function renders these HTML templates and passes data to them, allowing for dynamic web content generation.

---

### Q5. Create a Simple API and Test with Postman
'''
from flask import Flask, jsonify

app = Flask(__name__)

@app.route('/api/greet', methods=['GET'])
def greet():
    return jsonify({"message": "Hello from the API!"})

if __name__ == '__main__':
    app.run()
```
